```python
import matplotlib.pyplot as plt

# 坐标点
x1, y1 = 2, 3

# 绘制坐标系
plt.plot([0, x1], [y1, y1], 'k--')  # 画出水平虚线
plt.plot([x1, x1], [0, y1], 'k--')  # 画出垂直虚线

# 标出坐标点
plt.plot(x1, y1, 'ro')  # 'ro'表示红色圆点

# 添加坐标标签
plt.text(x1, y1, f'({x1}, {y1})', ha='right', va='bottom')  # 标注坐标

# 设置坐标轴范围
plt.xlim(0, 5)
plt.ylim(0, 5)

# 显示网格
plt.grid(True)

# 显示图形
plt.show()
```

```python
import numpy as np


def ss(s, t, i):
    mid = (s + t) / 2
    return np.concatenate([np.flipud(np.arange(mid, s - i * 0.5, -i)[1:]), np.arange(mid, t + i * 0.5, i)])


ss(1, 2 + 0.000001, 0.1)
```
